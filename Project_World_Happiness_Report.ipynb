```python
import pandas as pd
```

```python
#Se carga la data
df_2015 = pd.read_csv('2015.csv')
df_2016 = pd.read_csv('2016.csv')
df_2017 = pd.read_csv('2017.csv')
df_2018 = pd.read_csv('2018.csv')
df_2019 = pd.read_csv('2019.csv')
#Visualizamos la data verificando por el momento las dos primeras lineas, observando que no hay unificacin en los nombres de las columnas
display(df_2015.head(2), df_2016.head(2), df_2017.head(2), df_2018.head(2), df_2019.head(2))
```

```python
#Visualizamos las columnas por ao con sus respectivos nombres para verificar su uniformidad
print("Columnas de 2015:", df_2015.columns)
print("Columnas de 2016:", df_2016.columns)
print("Columnas de 2017:", df_2017.columns)
print("Columnas de 2018:", df_2018.columns)
print("Columnas de 2019:", df_2019.columns)
```

```python
#Al identificar los nombres de columnas con alguna diferencia se busca su uniformidad realizando su renombramiento y uniformidad

# Renombrar columnas en 2015
df_2015.rename(columns={
    'Economy (GDP per Capita)': 'GDP per Capita',
    'Health (Life Expectancy)': 'Healthy Life Expectancy',
    'Trust (Government Corruption)': 'Perceptions of Corruption',
    'Family': 'Social Support',
    'Freedom': 'Freedom to Make Life Choices'
}, inplace=True)

# Renombrar columnas en 2016
df_2016.rename(columns={
    'Economy (GDP per Capita)': 'GDP per Capita',
    'Health (Life Expectancy)': 'Healthy Life Expectancy',
    'Trust (Government Corruption)': 'Perceptions of Corruption',
    'Family': 'Social Support',
    'Freedom': 'Freedom to Make Life Choices'
}, inplace=True)

# Renombrar columnas en 2017
df_2017.rename(columns={
    'Happiness.Score': 'Happiness Score',
    'Economy..GDP.per.Capita.': 'GDP per Capita',
    'Health..Life.Expectancy.': 'Healthy Life Expectancy',
    'Trust..Government.Corruption.': 'Perceptions of Corruption',
    'Family': 'Social Support',
    'Freedom': 'Freedom to Make Life Choices'
}, inplace=True)

# Renombrar columnas en 2018 y 2019
df_2018.rename(columns={
    'Country or region': 'Country',
    'Score': 'Happiness Score',
    'GDP per capita': 'GDP per Capita',
    'Social support': 'Social Support',
    'Healthy life expectancy': 'Healthy Life Expectancy',
    'Freedom to make life choices': 'Freedom to Make Life Choices',
    'Perceptions of corruption': 'Perceptions of Corruption',
    'Overall rank': 'Happiness Rank'
}, inplace=True)

df_2019.rename(columns={
    'Country or region': 'Country',
    'Score': 'Happiness Score',
    'GDP per capita': 'GDP per Capita',
    'Social support': 'Social Support',
    'Healthy life expectancy': 'Healthy Life Expectancy',
    'Freedom to make life choices': 'Freedom to Make Life Choices',
    'Perceptions of corruption': 'Perceptions of Corruption',
    'Overall rank': 'Happiness Rank'
}, inplace=True)
```

```python
# Se agregar una columna llamada 'Year' a cada DataFrame para identificar el ao de suceso ya que la data no lo tiene.
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019
```

```python
#Se unifican o concatenan los 5 archivos en uno solo llamado df_all_years
df_all_years = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)
display(df_all_years.head())
```

```python
#Revisamos la informacin, como por ejemplo los no nulos, comparar columnas entre otros
display(df_all_years.info())
```

```python
#Revisamos la metricas estadisticas
display(df_all_years.describe())
```

```python
#Revisamos cuantos no nulos existen por columna
display(df_all_years.isnull().sum())
```

```python
#Revisamos cuantos no nulos existen por columna pero esta vez de forma grafica
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df_all_years.isnull(), cbar=False, cmap='viridis')
plt.title("Mapa de calor de valores faltantes")
plt.show()
```

```python
#Se eliminan las columnas donde predominan los nulos y que se identificaron en el paso anterior.
df_cleaned = df_all_years.drop(columns=[
    'Region', 'Standard Error', 'Lower Confidence Interval', 'Upper Confidence Interval',
    'Happiness Rank', 'Happiness.Rank', 'Whisker.high', 'Whisker.low', 'Dystopia Residual', 'Dystopia.Residual'
])
```

```python
#Se eliminan filas con valores nulos
df_cleaned = df_cleaned.dropna(subset=[
    'Happiness Score', 'GDP per Capita', 'Social Support', 'Healthy Life Expectancy',
    'Freedom to Make Life Choices', 'Generosity', 'Perceptions of Corruption'
])
```

```python
#Se verifica que ya no existan valores nulos
print("Valores nulos restantes por columna:", df_cleaned.isnull().sum())
```

```python
#Revisamos como queda nuestro dataframe y con cuantas columnas contamos para seguir trabajando en ella
display(df_cleaned.head())
```

```python
print("Aos presentes en el DataFrame final:", df_cleaned['Year'].unique())
```

```python
#Realizamos una agrupacin de los aos y calculamos la media o promedio
happiness_by_year = df_cleaned.groupby('Year')['Happiness Score'].mean()
display(happiness_by_year)
```

```python
#Realizamos un grafico que muestre el comportamiento promedio de la variable felicidad al pasar de los aos

plt.figure(figsize=(10, 6))
plt.plot(happiness_by_year, marker='o', linestyle='--', color='darkblue')
plt.title("Evolucin del ndice de felicidad Promedio del 2015 al 2019", fontsize=14)
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Happiness Score Promedio', fontsize=12)
plt.grid(True)
plt.xticks(happiness_by_year.index)  # Asegura que todas las etiquetas del eje X se muestren
plt.show()
```

El anterior grafico refleja principalmente una caida en el ndice de felicidad en el ao 2017, lo cual habra que investigar mas a fondo y ver que otras variables incidieron en dicho suceso. Y, despues de dicho ao el ndice de felicidad fue subiendo y en el 2019 llego a su pico mas alto.



```python
#Realizamos un grafico que muestre el comportamiento promedio de la variable Esperanza de vida saludable al pasar de los aos

import matplotlib.pyplot as plt
import seaborn as sns

healthy_life_by_year = df_cleaned.groupby('Year')['Healthy Life Expectancy'].mean()

plt.figure(figsize=(10, 6))
plt.plot(healthy_life_by_year, marker='o', linestyle='-', color='green')
plt.title("Evolucin de la Esperanza de vida saludable Promedio del 2015 al 2019", fontsize=14)
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Healthy Life Expectancy Promedio', fontsize=12)
plt.grid(True)
plt.xticks(healthy_life_by_year.index)
plt.show()
```

El anterior grafico refleja igualmente una caida en la Esperanza de vida saludable en el ao 2017, y de igual forma habra que investigar mas a fondo y ver que otras variables incidieron en dicho suceso. Y, despues de dicho ao la Esperanza de vida saludable fue subiendo y en el 2019 llego a su pico mas alto.

```python
#Visualizamos y revisamos la correlacin que existe entre las variables

variables = ['Happiness Score', 'GDP per Capita', 'Social Support', 'Healthy Life Expectancy',
             'Freedom to Make Life Choices', 'Generosity', 'Perceptions of Corruption']
correlation_matrix = df_cleaned[variables].corr()
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 10)) # Aumenta el tamao de la figura
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", annot_kws={'size': 10})
plt.title("Matriz de Correlacin entre Variables", fontsize=14)
plt.xticks(rotation=90) # Rota etiquetas del eje x para mejor lectura
plt.yticks(rotation=0)
plt.show()
```

```python
#Realizamos un top 5 de los paises mas felices y menos felices en 2019

df_2019_clean = df_cleaned[df_cleaned['Year'] == 2019]
top_5_happiest = df_2019_clean.nlargest(5, 'Happiness Score')[['Country', 'Happiness Score']]
bottom_5_least_happy = df_2019_clean.nsmallest(5, 'Happiness Score')[['Country', 'Happiness Score']]
print("Pases ms felices en 2019:")
display(top_5_happiest)

print("Pases menos felices en 2019:")
display(bottom_5_least_happy)
```

```python
#Graficamos el top 5 de paises mas felices en 2019

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Happiness Score', y='Country', data=top_5_happiest, palette='viridis', hue='Country', legend=False)
plt.title("Top 5 Pases Ms Felices en 2019")
plt.xlabel('Happiness Score')
plt.ylabel('Pas')
plt.show()
```

```python
#Graficamos el top 5 de paises menos felices en 2019

plt.figure(figsize=(10, 6))
sns.barplot(x='Happiness Score', y='Country', data=bottom_5_least_happy, palette='inferno', hue='Country', legend=False)
plt.title("Top 5 Pases Menos Felices en 2019")
plt.xlabel('Happiness Score')
plt.ylabel('Pas')
plt.show()
```

```python
#Realizamos un histograma para ver graficamente la distribucin del Happiness Socore
plt.figure(figsize=(5, 4))
plt.hist(df_cleaned['Happiness Score'], bins=10, color='teal', edgecolor='black')
plt.title("Distribucin del Happiness Score - Histograma")
plt.xlabel('Happiness Score')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

En el anterior grafico se puede observar que la mayoria de paises tienen una felicidad media, segun la grafica los que estan entre 4 y 7, pocos paises con felidad baja que son los que aparecen en el numero 3 y que segun los anteriores analisis serian Rwanda, Tanzania, Afghanistan. Y, entre el 7 y 8 los paises con alta felicidad que de igual manera segun los anteriores analisis serian Finland, Denmark, Norway entre otros.

```python
gdp_happiness_corr = df_cleaned['GDP per Capita'].corr(df_cleaned['Happiness Score'])
print(f"Correlacin entre GDP per Capita y Happiness Score: {gdp_happiness_corr:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text
import pandas as pd

plt.figure(figsize=(14, 10))
sns.scatterplot(x='GDP per Capita', y='Happiness Score', data=df_cleaned, color='skyblue', s=50)
sns.regplot(x='GDP per Capita', y='Happiness Score', data=df_cleaned, scatter=False, color='red')
plt.title("Relacin entre GDP per Capita y Happiness Score")
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')

# Etiquetar algunos puntos (Top 5 ms felices y Top 5 menos felices)
top_bottom_countries = pd.concat([top_5_happiest['Country'], bottom_5_least_happy['Country']])
labeled_data = df_cleaned[df_cleaned['Country'].isin(top_bottom_countries)]

texts = []
for index, row in labeled_data.iterrows():
    texts.append(plt.text(row['GDP per Capita'], row['Happiness Score'], row['Country'], fontsize=9, ha='left', va='bottom'))

# Usar adjust_text para evitar la superposicin
adjust_text(texts, force_points=(0.2, 0.2), force_text=(0.2, 0.2),
            arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

plt.show()
```

Segun la anterior grafica podemos observar una correlacin alta entre el GPD Per Capita, que es el indicador econmico que muestra la riqueza generada por un pas por cada habitante y el Happiness Score (puntuacin de felicidad), vemos los paises nordicos (Filand, Iceland, etc) que al tener un gpd per capita alto su nivel de felicidad es alto, lo contrario con la parte baja del grafico con paises como Tanzania o Rwanda con un gpd per capita bajo y un nivel de felicidad bajo, es por ello de la correlacin, ya que al parecer esta ligado el nivel economico con la puntuacin o nivel de felicidad.
